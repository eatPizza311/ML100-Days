2. è³‡è¨Šå†—é¤˜: æ¯”å¦‚åªåœ¨ä¹æ˜¯å¦åŠæ ¼ï¼Œå€‹åˆ¥åˆ†æ•¸å°±ä¸å¤ªé‡è¦ï¼Œå°‡åˆ†æ•¸æ”¹ç‚º0å’Œ1å³å¯ï¼Œéœ€åšäºŒåˆ†åŒ–ã€‚
3. é¡åˆ¥å‹ç‰¹å¾µä¸èƒ½ç›´æ¥ä½¿ç”¨: éœ€å°‡å…¶ç·¨ç¢¼å¾Œæ‰èƒ½è¼¸å…¥æ¨¡å‹ä¸­ã€‚

- ä½¿ç”¨æœ€å°é‡çš„ç‰¹å¾µå·¥ç¨‹ä»¥åŠç·šæ€§å›æ­¸æ¨¡å‹åšé æ¸¬, æœ€å¾Œè¼¸å‡ºå¯ä»¥åœ¨Kaggleæäº¤çš„é æ¸¬æª”

````python
# è¼‰å…¥å¥—ä»¶
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

#è®€å–è¨“ç·´èˆ‡æ¸¬è©¦è³‡æ–™
data_path = '../Data/Part02/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')
df_test = pd.read_csv(data_path + 'house_test.csv.gz')
print(df_train.shape)
````

### è¨“ç·´è³‡æ–™éœ€è¦ train_X, train_Y / é æ¸¬è¼¸å‡ºéœ€è¦ ids(è­˜åˆ¥æ¯å€‹é æ¸¬å€¼), test_X
### åœ¨æ­¤å…ˆæŠ½é›¢å‡º train_Y èˆ‡ ids, è€Œå…ˆå°‡ train_X, test_X è©²æœ‰çš„è³‡æ–™åˆä½µæˆ df, å…ˆä½œç‰¹å¾µå·¥ç¨‹

å…¶ä¸­[numpy.log1p](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.log1p.html)ç‚ºè¨ˆç®— $ln(1 + x)$ï¼Œå¯ä»¥é¿å…å°æ•¸é»è¢«å¿½ç•¥çš„æƒ…æ³ï¼Œå¦‚ä¸‹
```
>>> 1e-100 == 0.0
False
>>> 1e-100 + 1.0 == 1.0
True
```
å¦å¤–ä¸»è¦çš„åŠŸèƒ½æ˜¯ä½¿æœ‰åæ…‹çš„è³‡æ–™åˆ†å¸ƒæ›´æ¥è¿‘é«˜æ–¯åˆ†å¸ƒï¼Œè®“é æ¸¬çµæœæ›´å¥½

åƒè€ƒ[What is the purpose of numpy.log1p( )?](https://stackoverflow.com/questions/49538185/what-is-the-purpose-of-numpy-log1p)

````python
train_Y = np.log1p(df_train['SalePrice'])
ids = df_test['Id']
df_train = df_train.drop(['Id', 'SalePrice'] , axis=1)
df_test = df_test.drop(['Id'] , axis=1)
df = pd.concat([df_train,df_test])
df.head()
````

###  ğŸ’› ç‰¹å¾µå·¥ç¨‹-ç°¡åŒ–ç‰ˆ : å…¨éƒ¨ç©ºå€¼å…ˆè£œ-1, æ‰€æœ‰é¡åˆ¥æ¬„ä½å…ˆåš LabelEncoder, ç„¶å¾Œå†èˆ‡æ•¸å­—æ¬„ä½åšMinMaxScaler
### é€™å€å¡Šçš„ç´°ç¯€æœƒåœ¨å¾ŒçºŒçš„èª²ç¨‹åšè§£èªª

å…¶ä¸­list()å’Œ.reshape(-1, 1)éƒ½æ˜¯ç‚ºäº†è¦èª¿æ•´æ ¼å¼ï¼Œå› ç‚ºæ ¼å¼å’ŒScalerè¦æ±‚çš„ä¸åŒå°±æœƒå‡ºéŒ¯

1. list: æ•¸å€¼å‹çš„arrayæ˜¯æœ‰æ˜ç¢ºç¶­åº¦çš„, æ‰€ä»¥æ•¸å€¼å‹æ¬„ä½.valueså¾Œå¯ä»¥ç•¶è¼¸å…¥ï¼Œè€Œæ–‡å­—å‹çš„arrayå‰‡ç¶­åº¦æœ‰æ¨¡ç³Šåœ°å¸¶ (æœ‰å¹¾å€‹å­—? é‚„æ˜¯å­—è£¡é¢æœ‰å¹¾å€‹å­—æ¯?)ï¼Œæ‰€ä»¥è¦ç”¨listé™åˆ¶ä½, ä¸¦è®“scalerçŸ¥é“ä»–çš„ç¶­åº¦
2. reshape: MinMaxScalerè¼¸å…¥åªåƒäºŒç¶­é™£åˆ—ï¼Œå› æ­¤ä¸€ç¶­é™£åˆ—è¦é€²å», å°±å¿…é ˆå…ˆReshapeæˆäºŒç¶­é™£åˆ—æ‰èƒ½è¼¸å…¥ï¼Œ(-1, 1)ä»£è¡¨ä¸è«–å¹¾rowsï¼Œä½†é™åˆ¶è¦ä¸€å€‹column

````python
LEncoder = LabelEncoder()
MMEncoder = MinMaxScaler()
for c in df.columns:
    df[c] = df[c].fillna(-1)
    if df[c].dtype == 'object':
        df[c] = LEncoder.fit_transform(list(df[c].values))
    # ç„¡é‡ç¶±åŒ– - å€é–“ç¸®æ”¾    
    df[c] = MMEncoder.fit_transform(df[c].values.reshape(-1, 1))
df.head()
````

###  ğŸ’› ç‰¹å¾µå·¥ç¨‹-ç°¡åŒ–ç‰ˆ :å°‡å‰è¿°è½‰æ›å®Œç•¢è³‡æ–™ df , é‡æ–°åˆ‡æˆ train_X, test_X

````python
train_num = train_Y.shape[0]
train_X = df[:train_num]
test_X = df[train_num:]
````

### ä½¿ç”¨ç·šæ€§è¿´æ­¸æ¨¡å‹
### train_X, train_Y è¨“ç·´æ¨¡å‹, ä¸¦å° test_X åšå‡ºé æ¸¬çµæœ pred

````python
from sklearn.linear_model import LinearRegression
estimator = LinearRegression()
estimator.fit(train_X, train_Y)
pred = estimator.predict(test_X)
````

### å°‡è¼¸å‡ºçµæœ pred èˆ‡å‰é¢ç•™ä¸‹çš„ ID(ids) åˆä½µ, è¼¸å‡ºæˆæª”æ¡ˆ
### å¯ä»¥ä¸‹è¼‰ä¸¦é»é–‹ house_baseline.csv æŸ¥çœ‹çµæœ, ä»¥ä¾¿äº†è§£é æ¸¬çµæœçš„è¼¸å‡ºæ ¼å¼
### æœ¬ç¯„ä¾‹æ‰€èˆ‡ä½œæ¥­æ‰€è¼¸å‡ºçš„ csv æª”, å‡å¯ç”¨æ–¼æœ¬é¡Œçš„ Kaggle ç­”æ¡ˆä¸Šå‚³, å¯ä»¥è©¦è‘—ä¸Šå‚³ä¾†ç†Ÿæ‚‰ Kaggle çš„ä»‹é¢æ“ä½œ

````python
# å‰é¢åšäº†log1p æ‰€ä»¥è¦é‚„åŸ
pred = np.expm1(pred)
sub = pd.DataFrame({'Id': ids, 'SalePrice': pred})
sub.to_csv('house_baseline.csv', index=False) 
````

# ä½œæ¥­ : (Kaggle)éµé”å°¼ç”Ÿå­˜é æ¸¬ç²¾ç°¡ç‰ˆ

# ä½œæ¥­1
* ä¸‹åˆ—A~Eäº”å€‹ç¨‹å¼å€å¡Šä¸­ï¼Œå“ªä¸€å¡Šæ˜¯ç‰¹å¾µå·¥ç¨‹?
> å€å¡ŠC

# ä½œæ¥­2
* å°ç…§ç¨‹å¼å€å¡Š B èˆ‡ C çš„çµæœï¼Œè«‹å•é‚£äº›æ¬„ä½å±¬æ–¼"é¡åˆ¥å‹æ¬„ä½"? (å›ç­”æ¬„ä½è‹±æ–‡åç¨±å³å¯) 
> Name, Sex, Ticket, Cabin, Embarked

# ä½œæ¥­3
* çºŒä¸Šé¡Œï¼Œè«‹å•å“ªå€‹æ¬„ä½æ˜¯"ç›®æ¨™å€¼"?
> Survived

````python
# ç¨‹å¼å€å¡Š A
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data_path = '../Data/Part02/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')
df_train.shape
````

````python
# ç¨‹å¼å€å¡Š B
train_Y = df_train['Survived']
ids = df_test['PassengerId']
df_train = df_train.drop(['PassengerId', 'Survived'] , axis=1)
df_test = df_test.drop(['PassengerId'] , axis=1)
df = pd.concat([df_train,df_test])
df.head()
````

````python
# ç¨‹å¼å€å¡Š C
LEncoder = LabelEncoder()
MMEncoder = MinMaxScaler()
for c in df.columns:
    df[c] = df[c].fillna(-1)
    if df[c].dtype == 'object':
        df[c] = LEncoder.fit_transform(list(df[c].values))
    df[c] = MMEncoder.fit_transform(df[c].values.reshape(-1, 1))
df.head()
````

````python
# ç¨‹å¼å€å¡Š D
train_num = train_Y.shape[0]
train_X = df[:train_num]
test_X = df[train_num:]

from sklearn.linear_model import LogisticRegression
estimator = LogisticRegression()
estimator.fit(train_X, train_Y)
pred = estimator.predict(test_X)
````

````python
# ç¨‹å¼å€å¡Š E
sub = pd.DataFrame({'PassengerId': ids, 'Survived': pred})
sub.to_csv('titanic_baseline.csv', index=False) 
````
