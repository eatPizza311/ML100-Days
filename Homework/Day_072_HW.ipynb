### 何謂激活函數
> 激活函數定義了每個節點（神經元）輸出和輸入的關係，為神經元提供規模化非線性化能力，讓神經網絡具備強大的擬合能力
### 輸出值的範圍
> - 當激活函數輸出值是有限的時候，基於梯度的優化方法會更加穩定，因為特徵的表示受有限權值的影響更顯著
- 當激活函數的輸出是無限的時候，模型的訓練會更加高效
### 激活函數的作用
> - 深度學習的基本原理是基於人工神經網絡，信號從一個神經元進入，經過非線性的activation function
    - 如此循環往復，直到輸出層。正是由於這些非線性函數的反復疊加，才使得神經網絡有足夠的 capacity 來抓取複雜的 pattern
- 激活函數的最大作用就是非線性化。
- 激活函數的一個重要特徵是，它應該是可微分的，以便在網絡中向後推進以計算相對於權重的誤差梯度時執行反向優化策略，然後相應地使用梯度下降或任何優化技術優化權重以減少誤差。

### - Sigmoid
神經網路最常用的激活函數之一，特點是把輸出限定在 0~1 之間，使得數據在傳遞過程中不容易發散，其計算公式為:
$$sigmoid(x)\ =\ \frac{1}{1+e^{-x}}$$
#### Sigmoid 導數
$$(sigmoid(x))'\ =\ \frac{e^{-x}}{(1+e^{-x})^2}\ =\ sigmoid(x)(1-sigmoid(x))$$
#### 缺點:
1. 容易過飽和(限制輸出範圍)，丟失梯度。這樣在反向傳播時，很容易出現梯度消失的情況，導致訓練無法完整
2. 輸出均值不是 0
#### 適用範圍:
> 主要用於二元分類的輸出層

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

x = np.arange(-10, 10, 0.01)
y = 1/(1+np.exp(-x))
dy = np.exp(-x)/(1+np.exp(-x))**2


# sigmoid figure
# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, y, 'b', label='linspace(-10,10,10)')
plt.plot(x, dy, 'g', label='linespace=(-10, 10, 10)')


# Draw the grid line in background.
plt.grid()

# Title
plt.title('Sigmoid Function')

# write the Sigmoid formula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)
plt.text(3, 0.2, r"$\sigma(x)' = \sigma(x)(1-\sigma(x)$", fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# create the graph
plt.show()
```

### - Softmax
Softmax 把一個 k 維的 real value 向量（a1,a2,a3,a4....）映射成一個（b1,b2,b3,b4....）其中 bi 是一個 0～1 的常數，輸出神經元之和為1.0，所以可以拿來做多分類的機率預測，其計算公式為:

$$\sigma(z)_j\ =\ \frac{e^z_j}{\sum^K_{k=1}{e^z_k}}$$

#### 為什麼要取指數?
- 第一個原因是要模擬 max 的行為，所以要讓大的更大。
- 第二個原因是需要一個可導的函數。

```python
x = np.arange(-5, 5, 0.1)
y = np.exp(x) / float(sum(np.exp(x)))

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))

print(y)
plt.plot(x, y, 'r')
plt.show()
```

### - tanh
tanh 也稱為雙切正切函數，取值範圍為 [-1,1]，在特徵相差明顯時的效果會很好，在循環過程中會不斷擴大特徵效果，其計算公式為:

$$tanh(x)\ =\ \frac{e^x-e^{-x}}{e^x+e^{-x}}$$

#### tanh 導數
$$tanh(x)'\ =\ 1-tanh(x)^2$$

#### 適用範圍
> 幾乎適用於任何情況

```python
x = np.arange(-10, 10, 0.01)
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
plt.plot(x,tanh(x))
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.plot(x, 1-tanh(x)**2, 'g')

# Title
plt.title('tanh')

# write the Sigmoid formula
plt.text(4, 0.8, r'$\tanh(x)}$', fontsize=15)
plt.text(3, 0.2, r"$\tanh'(x)$", fontsize=15)
plt.show()
```

## Sigmoid vs Softmax 
- Sigmoid 將一個 real value 映射到（0,1）的區間，用來做二分類
- Softmax 把一個 k 維的 real value 向量（a1,a2,a3,a4....）映射成一個（b1,b2,b3,b4....）其中 bi 是一個 0～1 的常數，輸出神經元之和為 1.0，所以可以拿來做多分類的機率預測二分類問題時 
- sigmoid 和 softmax 是一樣的，求的都是 cross entropy loss

## Sigmoid vs Tanh
- tanh 函數將輸入值壓縮到 -1~1 的範圍，因此它是 0 均值的，解決了Sigmoid 函數的非 zero-centered 問題，但是它也存在梯度消失和冪運算的問題。
- 其實 $tanh(x)=2sigmoid(2x)-1$

### - Rectified Linear Unit- Relu 
修正線性單元（Rectified linear unit，ReLU），其計算公式為:
$$f(x)=max(0,x)$$
- 在 x > 0 時導數恆為1
- 對於 x < 0，其梯度恆為 0，這時候它也會出現飽和的現象，甚至使神經元直接無效，從而其權重無法得到更新(在這種情況下通常稱為 dying ReLU)
- Leaky ReLU 和 PReLU 的提出正是為了解決這一問題

#### ReLU 導數
$$f'(x)=\left\{\begin{matrix}1,\ x\geq0
\\ 0,\ x<0
\end{matrix}\right. $$
#### 適用範圍
> 二分類問題選擇sigmoid，剩下的問題ReLU是默認的選擇。

```python
'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
x = np.arange(-10, 10, 0.01)
y = list(map(lambda x: x if x>0 else 0, x))
dy = list(map(lambda x: 1 if x > 0 else 0, x))
plt.plot(x, y)
plt.plot(x, dy, 'g')
plt.text(3, 8, 'ReLU', fontsize=15)
plt.text(3, 1.5, 'ReLU derivative', fontsize=15)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.show()
```

#### ELU (exponential linear unit)
- ELU 函數是針對 ReLU 函數的一個改進型，相比於 ReLU 函數，在輸入為負數的情況下，是有一定的輸出的，其計算公式為:

$$f(x)=\left\{\begin{matrix}x,\ x>0
\\ \alpha(e^x-1)),\ x\leq0
\end{matrix}\right.$$

- 這樣可以消除 ReLU 死掉的問題
- 還是有梯度飽和和指數運算的問題

```python
x = np.arange(-10, 10, 0.01)
a = 0.8
y = list(map(lambda x: x if x>0 else a*(np.exp(x)-1), x))
plt.plot(x, y)
plt.text(3, 8, 'ELU', fontsize=15)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.show()
```

#### PReLU 
參數化修正線性單元（Parameteric Rectified Linear Unit，PReLU）屬於 ReLU 修正類激活函數的一員，其計算公式為:

$$f(x)=max(\alpha x, x)$$

其中$\alpha \in [0, 1]$

#### Leaky ReLU
當 $\alpha = 0.1$ 我們叫 PReLU 為Leaky ReLU，算是 PReLU 的一種特殊情況，PReLU 以及 Leaky ReLU 有一些共同點，即爲負值輸入添加了一個線性項，這個線性項的斜率實際上是在模型訓練中學習到的。

```python
x = np.arange(-10, 10, 0.01)
a = 0.1
y = list(map(lambda x: x if x>0 else a*x, x))
plt.plot(x, y)
plt.text(2, 8, 'Leaky ReLU', fontsize=15)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
plt.show()
```

### - Maxout
Maxout 是深度學習網絡中的一層網絡，就像池化層、卷積層一樣，可以看成是網絡的激活函數層，會取得所有這些「函數層」中的最大值。Maxout 的擬合能力是非常強的，優點是計算簡單，不會過飽和，同時又沒有 ReLU 的缺點，但缺點是過程參數相當於多了一倍。其計算公式為:

$$f(x)\ =\ max(wT1x+b1,\ wT2x+b2)$$

## 如何選擇正確的激活函數
1. 根據各個函數的優缺點來配置
    - 如果使用 ReLU，要小心設置 learning rate，注意不要讓網絡出現很多「dead」神經元，如果不好解決，可以試試 Leaky ReLU、PReLU 或者Maxout。

2. 根據問題的性質
    - 用於分類器時，Sigmoid 函數及其組合通常效果更好。
    - 由於梯度消失問題，有時要避免使用 sigmoid 和 tanh 函數。ReLU 函數是一個通用的激活函數，目前在大多數情況下使用。
    - 如果神經網絡中出現死神經元，那麼 PReLU 函數就是最好的選擇。
    - ReLU 函數建議只能在隱藏層中使用

3. 考慮 DNN 損失函數和激活函數
    - 如果使用 sigmoid 激活函數，則交叉熵損失函數一般肯定比均方差損失函數好。
    - 如果是 DNN 用於分類，則一般在輸出層使用 softmax 激活函數。
    - ReLU 激活函數對梯度消失問題有一定程度的解決，尤其是在 CNN模型中。

### 梯度消失 Vanishing gradient problem
- 原因：前面的層比後面的層梯度變化更小，故變化更慢
- 結果：Output變化慢 → Gradient小 → 學得慢
- Sigmoid，Tanh都有這樣的特性，見下圖其導數左右兩端平緩的斜率即可了解，故不適合用在 Layers 多的DNN 架構

## Keras
激活函數可以通過設置單獨的激活層實現，也可在構造層對象時通過activation參數實現：
> from keras.layers  import  Activation,Dense <br>
model.add(Dense(64,activation=‘tanh’))

考慮不同Backend support，通過傳遞一個元素運算的Theano / TensorFlow / CNTK函數來作為激活函數：
> from keras import  backend as k <br>
model.add(Dense(64,activation=k.tanh))<br>
model.add( Activation(k.tanh))
